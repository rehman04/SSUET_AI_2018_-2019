```python
def fuzz_logic(x):
    max=0
    match=''
    for i in parties['Party']:
        r=f.ratio(x,i)
        if r>max:
            max=r
            match=i
    return match
```

```python
extracted_parties=eledata.Party.apply(fuzz_logic)
```
